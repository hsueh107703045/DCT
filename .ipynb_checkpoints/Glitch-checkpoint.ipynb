```python
import cv2
import numpy as np
import random
import time

time_start = time.time()

img_1 = cv2.imread('./images/tea1.png')
img_2 = cv2.imread('./images/tea2.png')

height = img_1.shape[0]
width = img_1.shape[1]

print(height, width)

if img_2.shape[0] != height or img_2.shape[1] != width:
    print("Different shape! (image 2 resized) ")
    img_2 = cv2.resize(img_2, (width, height))

img_out = np.zeros((height,width,3), np.uint8)

while True:
    for alpha in np.linspace(0, 1, 100):
        img_out = cv2.addWeighted(img_1, alpha, img_2, (1-alpha), gamma=0)
        #img_out = img_1 * alpha + img_2 * (1-alpha)
        #img_out = img_1 #why?????
        
        
        
        #'''
        ### shifting
        shift_range = np.round(np.random.rand(2) * height)
        #shift_range = np.sort(shift_range) # if not sorted, just skip
        #print(shift_range)
        
        shift_lengthB = random.randint(1, width//20)
        shift_lengthG = random.randint(1, width//20)
        shift_lengthR = random.randint(1, width//20)
        
        img_out[int(shift_range[0]):int(shift_range[1]),:,0] = np.roll(img_out[int(shift_range[0]):int(shift_range[1]),:,0], shift_lengthB)
        #img_out[0:80, :, 0] = np.roll(img_out[0:80, :,0], shift_lengthB)
        img_out[int(shift_range[0]):int(shift_range[1]),:,1] = np.roll(img_out[int(shift_range[0]):int(shift_range[1]),:,1], shift_lengthG)
        img_out[int(shift_range[0]):int(shift_range[1]),:,2] = np.roll(img_out[int(shift_range[0]):int(shift_range[1]),:,2], shift_lengthR)
        #'''

        
        
        
        #'''
        ### random value
        remove_rangeB = (np.round(np.random.rand(2) * height/10) + random.randint(0,9999)) % height
        remove_rangeG = (np.round(np.random.rand(2) * height/10) + random.randint(0,9999)) % height
        remove_rangeR = (np.round(np.random.rand(2) * height/10) + random.randint(0,9999)) % height

        img_out[int(remove_rangeB[0]):int(remove_rangeB[1]), :, 0] = random.randint(0, 255)
        img_out[int(remove_rangeG[0]):int(remove_rangeG[1]), :, 1] = random.randint(0, 255)
        img_out[int(remove_rangeR[0]):int(remove_rangeR[1]), :, 2] = random.randint(0, 255)
        #'''
        
        
        
        
        #'''
        ### block overflow
        overflow_corner = np.round(np.random.rand(2) * height)  # can't reach whole pic
        overflow_size = np.round(np.random.rand(2) * height//5)
        #print(overflow_range)
        img_out[int(overflow_corner[0]):int(overflow_corner[0]+overflow_size[0]), int(overflow_corner[1]):int(overflow_corner[1]+overflow_size[1]), 0] += 200
        img_out[int(overflow_corner[0]):int(overflow_corner[0]+overflow_size[0]), int(overflow_corner[1]):int(overflow_corner[1]+overflow_size[1]), 1] += 200
        img_out[int(overflow_corner[0]):int(overflow_corner[0]+overflow_size[0]), int(overflow_corner[1]):int(overflow_corner[1]+overflow_size[1]), 2] += 200
        #'''
        

        #'''
        ### add noise   # use several precomputed noise maps?? faster? OR use one noise map and np.roll() it every time 
        mean = 0
        sigma = 0.1
        noise = np.random.normal(mean, sigma, img_out.shape)
        img_out = img_out / 255
        img_out = img_out + noise
        img_out = np.clip(img_out, 0, 1)
        img_out = np.uint8(img_out*255)
        #'''



        cv2.imshow('out', img_out)
        cv2.waitKey(20)
    break


    

    
print("breaked")


time_end = time.time()
time_c= time_end - time_start
print('time cost', time_c, 's')




cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1) 



```
